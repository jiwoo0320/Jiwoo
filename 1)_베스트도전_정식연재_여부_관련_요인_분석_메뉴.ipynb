## **1) 장르별 정식연재 작품 여부**

```python
import csv
import matplotlib.pyplot as plt
f = open(directory_path + 'naver_challenge.csv', 'r')
data = csv.reader(f)
next(data)

# 장르별로 정식연재 작품과 전체 작품 수를 저장할 딕셔너리
genre_published_count = {}
genre_total_count = {}

# 데이터 수집
for row in data:
    genre = row[5].strip()  # 장르 저장
    published = row[-3].strip().lower() == 'true'  # 정식연재여부 확인 후 정식연재 작품 저장

    # 해당 장르일 경우 1씩 더해서 장르의 개수를 계속 업데이트해 해당 장르의 전체 작품 수를 집계
    genre_total_count[genre] = genre_total_count.get(genre, 0) + 1   # 만약 해당 장르가 없으면 기본값은 0

    # 정식연재 작품인 경우의 위와 마찬가지로 해당 장르의 정식연재 작품 수를 집계
    if published:
        genre_published_count[genre] = genre_published_count.get(genre, 0) + 1  # 만약 해당 장르가 없으면 기본값은 0

# 각 장르별 정식연재 비율을 계산해서 genre_ratios 딕셔너리에 저장
genre_ratios = {}
for genre, published_count in genre_published_count.items():
    total_count = genre_total_count.get(genre, 0)
    if total_count > 0:        # 분모가 0일 수 없기 때문
        ratio = published_count / total_count
        genre_ratios[genre] = ratio

#  딕셔너리의 키(장르)와 값(정식 연재 비율)을 사용하여 막대 그래프 생성
plt.rc('font', family='nanumBarunGothic')
plt.figure(figsize=(12, 6), dpi=300)
plt.bar(genre_ratios.keys(), genre_ratios.values(), color='green')
plt.title('장르-정식연재여부 간의 비율')
plt.xlabel('장르')
plt.ylabel('정식연재비율')
plt.xticks(rotation=45, ha='right')  # x축 레이블을 45도 회전해 오른쪽 정렬
plt.show()
```

## **2) 정식연재된 작품과 그렇지 않은 작품의 별점 비율 비교**

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
# CSV 파일 열기 및 읽기
f = open(directory_path + 'naver_challenge.csv', 'r')
data = csv.reader(f)
next(data)

# 정식연재된 작품과 정식연재되지 않은 작품의 개수를 저장할 변수 초기화
published_ratings = []
unpublished_ratings = []

# 데이터 수집
for row in data:
        published = row[-3].strip().lower() == 'true'   # 정식연재여부 확인 후 정식연재 작품 저장
        rating = float(row[4])   # 실수형으로 변환해 평점 저장
        # 정식연재 여부에 따라 평점 분류
        if published:
            published_ratings.append(rating)
        else:
            unpublished_ratings.append(rating)

# 별점 범주 정의
categories = [8.5, 9.0, 9.5, 9.8, 10]

# 별점 범주에 따라 정식연재작품 평점과 정식연재되지 않은 작품 평점 분류
published_counts, _ = np.histogram(published_ratings, bins=categories)  #  bin의 경계값은 '-'로 무시
unpublished_counts, _ = np.histogram(unpublished_ratings, bins=categories)    #  bin의 경계값은 '-'로 무시

# 그래프 그리기
plt.figure(figsize=(10, 6), dpi=300)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
color = ['lightgreen', 'orange', 'yellow', 'green']   #  각각의 범주를 서로 다른 색으로 지정

# 파이 차트 그리기 - 정식연재된 작품
plt.subplot(1, 2, 1)
plt.title('정식연재된 작품의 별점 비율')
plt.pie(published_counts,
        # 마지막 범주를 제외한 각 범주의 경계 값으로 레이블 지정
        labels=[f'{categories[i]}~{categories[i+1]}' for i in range(len(categories)-1)],
         # 마지막 레이블인 9.8-10을 0.05만큼 돌출시키고 소수점 첫번째 자리까지 표시
        colors=color, explode=(0,0,0,0.05), autopct='%1.1f%%', startangle=90)
plt.axis('equal')

# 파이 차트 그리기 - 정식연재되지 않은 작품
plt.subplot(1, 2, 2)
plt.title('정식연재되지 않은 작품의 별점 비율')
plt.pie(unpublished_counts,
        # 마지막 범주를 제외한 각 범주의 경계 값으로 레이블 지정
        labels=[f'{categories[i]}~{categories[i+1]}' for i in range(len(categories)-1)],
        # 마지막 레이블인 9.8-10을 0.05만큼 돌출시키고 소수점 첫번째 자리까지 표시
        colors=color, explode=(0,0,0,0.05),autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.legend()
plt.show()
```

## **3) 사용자가 입력한 장르에 따른 작품별 별점 분포**

```python
import csv
import matplotlib.pyplot as plt
f = open(directory_path + 'naver_challenge.csv', 'r')
data = csv.reader(f)
next(data)

# 사용자 입력 받기
user_genre = input("장르를 입력하세요: ")

# 장르별 별점을 저장할 리스트
ratings = []
titles = []

# 사용자가 입력한 장르에 해당하는 작품들의 별점을 수집
for row in data:
    genre = row[5].strip()
    if genre == user_genre:
            rating = float(row[4])  # 별점 데이터 실수형으로 가져오기
            titles.append(row[1])   # 작품 이름 추가
            ratings.append(rating)  # 별점 추가

if titles and ratings:  # 수집된 제목과 별점이 있는 경우
   # 산점도 그리기
   plt.rc('font', family='nanumBarunGothic')
   plt.rcParams.update({'font.size': 8})
   plt.rcParams['axes.unicode_minus']= False
   plt.figure(figsize=(10, 6), dpi=300)
   plt.scatter(titles, ratings, c=ratings, cmap='jet', alpha=0.7)
   plt.colorbar()
   plt.title(f'{user_genre} 장르 작품의 별점 분포')
   plt.xlabel('작품 이름')
   plt.ylabel('별점')
   plt.xticks(rotation=90, ha='right')
   plt.show()
else: # 사용자가 입력한 장르에 해당하는 작품이 없는 경우
    print(f"{user_genre} 장르에 해당하는 작품이 없습니다.")
```

## **4) 사용자가 입력한 키워드가 설명에 포함된 작품의 정식연재여부**

```python
import csv
import numpy as np
# CSV 파일 열기 및 읽기
f = open(directory_path + 'naver_challenge.csv', 'r')
data = csv.reader(f)
next(data)

# 사용자 입력 받기
user_keyword = input("검색할 키워드를 입력하세요: ")

# 정식연재된 작품과 정식연재되지 않은 작품의 개수를 저장할 변수 초기화
published_count = 0
unpublished_count = 0

# 데이터 수집
for row in data:
    summary = row[-4]   # 작품 설명 저장
    published = row[-3].strip().lower() == 'true' # 정식연재여부 확인 후 정식연재 작품 저장

    if user_keyword in summary.lower():  # 요약 데이터에 키워드가 포함되어 있는지 확인
        if published:
            published_count += 1   # 정식연재 작품일 경우 published_count가 1씩 증가
        else:
            unpublished_count += 1 # 정식연재되지 않은 작품일 경우 unpublished_count가 1씩 증가

# 결과 출력
print(f"'{user_keyword}'가 포함된 정식연재된 작품의 개수: {published_count}")
print(f"'{user_keyword}'가 포함된 정식연재되지 않은 작품의 개수: {unpublished_count}")

counts = np.array([published_count, unpublished_count])  # 넘파이 배열로 변환해서 하나의 배열로 합산
total_published = np.sum(counts)  # 정식연재 작품과 정식연재되지 않은 작품의 총합

# 상관관계 분석
if total_published > 0:  # 사용자가 입력한 키워드가 있을 경우
    ratios = counts / total_published * 100   # 총 작품 중 정식연재된 작품과 정식연재되지 않은 작품 각각의 비율 계산
    print(f"정식연재된 작품 중 '{user_keyword}'를 포함한 작품 비율: {ratios[0]:.2f}%")  # 소숫점 두번째 자리까지 표시
    print(f"정식연재되지 않은 작품 중 '{user_keyword}'를 포함한 작품 비율: {ratios[1]:.2f}%")   # 소숫점 두번째 자리까지 표시
else:   # 사용자가 입력한 키워드가 없을 경우
    print("해당 키워드를 포함한 작품이 존재하지 않습니다.")
```

## **5) 사용자가 입력한 키워드가 설명에 포함된 작품과 별점 간의 관계**

```python
import csv
import matplotlib.pyplot as plt
f = open(directory_path + 'naver_challenge.csv', 'r')
data = csv.reader(f)
next(data)

# 사용자 입력 받기
user_keyword = input("검색할 키워드를 입력하세요: ")

# 필터링된 작품들의 별점 저장을 위한 리스트 초기화
ratings = []

for row in data:
    summary = row[-4]  # 설명 저장
    rating = row[4]    # 별점 저장
    # 키워드가 설명에 포함되었다면 별점 데이터 실수형으로 변환해서 추가
    if user_keyword.lower() in summary.lower():
            ratings.append(float(rating))

# 필터링된 작품들의 개수 출력
print(f"'{user_keyword}'를 포함한 작품의 개수: {len(ratings)}")

# 별점과 키워드의 상관관계 계산
if ratings:   # 별점이 있다면
    keyword_count = sum(1 for r in ratings if r >= 9.8)  # 키워드 포함된 작품 중 별점이 7 이상인 작품의 수
    correlation = keyword_count / len(ratings) *100  # 키워드 포함된 작품 중 별점이 7 이상인 작품의 비율
    print(f"'{user_keyword}'에서 별점 9.8 이상인 작품 비율: {correlation:.2f}%")  # 소수점 둘째자리까지 표시
else:    # 별점이 없다면
    print("'{user_keyword}'을 포함한 작품 중 별점 9.8이상인 작품이 없습니다.")
```
