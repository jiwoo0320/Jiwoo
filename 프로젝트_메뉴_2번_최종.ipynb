# **[단계3: 경력 작가와 신인 작가의 평균별점 비교]**

**[작가명을 입력하면 별점을 알려주는 코드]**

```python
# 작가명을 입력하면 평균별점을 알려주는 코드
# 경력 작가일 경우 여러 개의 별점 데이터를 한번에 반환함
import csv
f = open(directory_path + 'webtoon.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')

writer = input("작가명 입력(예: Seokwoo, Yaongyi 등):")
for row in data :
    if writer in row[2] :   # row[2]에 입력한 writer가 존재하면
        print(row[5])       # 그 작가의 별점 데이터인 row[5]를 출력한다
```

**[신인작가와 경력작가를 구분하여 저장하는 리스트]**

```python
#신인 작가와 경력 작가를 구분하는 리스트
import csv
f = open(directory_path + 'webtoon.csv', 'r', encoding='utf-8')
data = csv.reader(f)
next(data)
new = []            # 신인 작가를 저장할 리스트
experienced = []    # 경력 작가를 저장할 리스트

for row in data:
    row[2] = str(row[2])        # 작가명 데이터를 문자열로 변환한다
    if row[2] == '':            # 작가명이 누락되어 빈칸인 경우 패스한다
        pass
    if row[2] == 'Various Artists':     # 여러 작가가 참여하여 'Various Artists'라고 저장된 경우, 이는 여러 번 존재한다고 해도 경력 작가가 아니므로 삭제함
        del row[2]
    else:
        if row[2] not in new:       # 신인 작가 리스트에 작가명이 없으면, 추가한다
            new.append(row[2])
        else:
            new.remove(row[2])      # 신인 작가 리스트에 이미 작가명이 있으면, 그 작가명을 지우고 경력 작가 리스트에 추가한다
            experienced.append(row[2])

print('new =', new)
print('experienced =', experienced)
```

```python
# new와 experienced에 겹치는 원소를 정리하는 carrer 함수

def career(a, b):
    for i in a[:]:
        if i in b:          # a 리스트 속 i가 b 리스트에도 존재하면
            a.remove(i)     # a 리스트에서 i를 삭제한다
    return a, b

career(new, experienced)    # 신인작가(new) 리스트 속 원소가 경력작가(experienced) 리스트에도 존재하면, 신인작가 리스트에서 이를 삭제한다
print('new =', new)
print('experienced =', experienced)
```

**[신인작가와 경력작가를 구분한 별점 리스트]**

```python
# 각각의 별점 리스트
import csv
f = open(directory_path + 'webtoon.csv', 'r', encoding='utf-8')
data = csv.reader(f)
next(data)

new_rate = []           # 신인 작가의 별점리스트
experienced_rate = []   # 경력 작가의 별점리스트

for row in data:
    if row[2] in new:                    # 만약 작가명이 신인 작가 리스트에 있다면
        new_rate.append(row[5])          # 그 별점을 new_rate에 추가함
    if row[2] in experienced:            # 만약 작가명이 경력 작가 리스트에 있다면
        experienced_rate.append(row[5])  # 그 별점을 experienced_rate에 추가함
    else:
        pass

print('new_rate =', new_rate)
print('experienced_rate =', experienced_rate)
```

```python
#리스트 내의 원소를 실수형으로 변환

for i in range(len(new_rate)):                  # 신인 작가 별점 리스트 속 길이만큼 확인한다
    if type(new_rate[i]) == str:                # 만약 신인 작가 별점 리스트 속 원소가 문자열이라면
        new_rate[i] = float(new_rate[i])        # 이를 실수형으로 변환함

print('new_rate 실수형 =', new_rate)

for i in range(len(experienced_rate)):                      # 경력 작가 별점 리스트 속 길이만큼 확인한다
    if type(experienced_rate[i]) == str:                    # 만약 경력 작가 별점 리스트 속 원소가 문자열이라면
        experienced_rate[i] = float(experienced_rate[i])    # 이를 실수형으로 변환함

print('experienced_rate 실수형 =', experienced_rate)
```

```python
#각각 rate의 평균을 세는 mean 함수
def mean(lst):
    if len(lst) == 0 :               # 리스트의 총 원소 개수가 0인 경우
        return 0                     # 0을 반환한다
    else:
        return sum(lst) / len(lst)   # 리스트의 총 원소 개수가 0이 아니라면, 합을 개수로 나눔

new_mean = mean(new_rate)                   # 신인 작가 별점 리스트의 평균
experienced_mean = mean(experienced_rate)   # 경력 작가 별점 리스트의 평균

print('신입 평균 =', new_mean)
print('경력직 평균 =', experienced_mean)
```

**결과: 신인 작가의 별점 평균이 경력 작가의 별점 평균에는 큰 차이가 없으며, 오히려 가설과 달리 신인 작가 측의 데이터가 살짝 높다**

# **[단계 4: 경력 작가와 신인 작가 평균 별점 관련 그래프 시각화]**

```python
# 별점 그래프 시각화를 위해 값이 작은 순서대로 다시 정렬
new_rate = sorted(new_rate)
experienced_rate = sorted(experienced_rate)
```

```python
# 신규 작가 그래프 시각화하기
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.title('신규 작가의 평균별점 분포')
plt.plot(new_rate)
plt.ylabel('평균별점')
plt.show()
```

```python
# 경력 작가 그래프 시각화하기
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.title('경력직 작가의 평균별점 분포')
plt.plot(experienced_rate)
plt.ylabel('평균별점')
plt.show()
```

**평균 별점에는 큰 차이가 없는 것으로 확인되었다.**
